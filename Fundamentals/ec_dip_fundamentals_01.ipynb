The image above has a relatively low dynamic range (DEFINE DYN RANGE). To enhance its visualization, we can threshold the image. 

In Matlab, the imshow() function has an autorange feature that thresholds the image given input threshold values. We can accomplish the same task in matplotlib by passing in the vmin and vmax parameters

First, let's try to threshold by the min and max values of the image to see what we get...

```python
#displaying autoranged image
body_fig2, body_ax2 = plt.subplots(1,1)
body_ax2.imshow(img_body_scan, cmap = 'gray', vmin = np.min(img_body_scan), vmax = np.max(img_body_scan))
```

Note that the dynamic range of the image did not improve substantially. To verify why this is the case, let's print out the min and max intensity values

```python
#printing min and max intensity values of our grayscale image
print("min intensity value is: ", np.min(img_body_scan))
print("max intensity value is: ", np.max(img_body_scan))
```

Since our intensity values span the entire 2^8 -1 possible pixel values, thresholding by 0 and 255 does not improve our displayed results.

## Writing Images

From OpenCV's reference page (accessed 1/22/21): https://docs.opencv.org/3.4/d4/da8/group__imgcodecs.html

**imwrite()**

Saves an image to a specified file.

The function imwrite saves the image to the specified file. The image format is chosen based on the filename extension (see cv::imread for the list of extensions). In general, only 8-bit single-channel or 3-channel (with 'BGR' channel order) images can be saved using this function, with these exceptions:

- 16-bit unsigned (CV_16U) images can be saved in the case of PNG, JPEG 2000, and TIFF formats
- 32-bit float (CV_32F) images can be saved in TIFF, OpenEXR, and Radiance HDR formats; 3-channel (CV_32FC3) TIFF images will be saved using the LogLuv high dynamic range encoding (4 bytes per pixel)
- PNG images with an alpha channel can be saved using this function. To do this, create 8-bit (or 16-bit) 4-channel image BGRA, where the alpha channel goes last. Fully transparent pixels should have alpha set to 0, fully opaque pixels should have alpha set to 255/65535 (see the code sample below).
- Multiple images (vector of Mat) can be saved in TIFF format (see the code sample below).

If the format, depth or channel order is different, use Mat::convertTo and cv::cvtColor to convert it before saving. Or, use the universal FileStorage I/O functions to save the image to XML or YAML format.

```python
#For our case, we have an 8-bit single channel image. Let's save it as a tiff file for consistency with the original format and reload it to check it out
#saving clipped image
cv2.imwrite("./images/partial_body_scan_clipped.tif",img_body_clipped)
```

## Changing the size of an image (tiff) without changing its resolution by manipulating the dots per inch (dpi) resolution

Example 2.3 fro DIPUM - the original image has a resolution of 2048 x 2048 pixels at 800 dpi --> its size is 2.56 x 2.56 in
 
We want to change the image size for printing from 2.56 x 2.56 in to 1.7 x 1.7 in without changing the number of pixels in the image. 

We can accomplish this by saving the image as a tiff file with a different dpi resolution for both the rows and columns

```python
#load test image pattern
img_test_pattern = cv2.imread('./images/testpattern2048.tif', cv2.IMREAD_UNCHANGED)

#plot original image
fig_tst_pttrn, ax_tst_pttrn = plt.subplots(1,1)
ax_tst_pttrn.imshow(img_test_pattern,cmap='gray'); ax_tst_pttrn.set_title('2048 x 2048 image');

#To resize image to 1.7 x 1.7 in, we need to modify our DPI as follows
res = np.round(800*((2.56/1.7)))
```

```python
#change dpi (resize image) while maintaining the same number of pixels
img_pttrn_rszed = cv2.imwrite('./images/output/testpattern2048_rszed.tif', img_test_pattern, [cv2.IMWRITE_TIFF_XDPI, res, cv2.IMWRITE_TIFF_YDPI, res])
```

```python
#plot both images
fig_tst_pttrn2, ax_tst_pttrn2 = plt.subplots(1,2)
ax_tst_pttrn2 = ax_tst_pttrn2.ravel()

#load resized image 
pttrn_rszed_img = cv2.imread('./images/output/testpattern2048_rszed.tif',cv2.IMREAD_UNCHANGED)

ax_tst_pttrn2[0].imshow(img_test_pattern,cmap='gray'); ax_tst_pttrn2[0].set_title('Original image');
ax_tst_pttrn2[1].imshow(pttrn_rszed_img,cmap='gray'); ax_tst_pttrn2[1].set_title('Resized image');
```

```python
#plot both images
orig_dpi = 800
pttrn_img_ht, pttrn_img_wd = img_test_pattern.shape

#what size does the figure need to be in inches to fit the image?
figsize1 = pttrn_img_wd / float(orig_dpi), pttrn_img_ht / float(orig_dpi)

#create a figure of the right size with one azes that takes up the full figure
pttrn_fig1 = plt.figure(figsize = figsize1)
ax1 = pttrn_fig1.add_axes([0,0,1,1])

##################
#resized image
##################

pttrn_img_ht2, pttrn_img_wd2 = pttrn_rszed_img.shape

#what size does the figure need to be in inches to fit the image?
figsize2 = pttrn_img_wd2 / float(res), pttrn_img_ht2 / float(res)

#create a figure of the right size with one azes that takes up the full figure
pttrn_fig2 = plt.figure(figsize = figsize2)
ax2 = pttrn_fig2.add_axes([0,0,1,1])

#display image
ax1.imshow(img_test_pattern,cmap='gray')
ax1.set_title('2.6 x 2.56 in Image')

#display image
ax2.imshow(pttrn_rszed_img,cmap='gray')
ax2.set_title('1.7 x 1.7 in Image')
```

```python
#Verify that both images have the same number of pixels
print('Original Image Info')
get_img_info(img_test_pattern)
print('\nResized Image Info')
get_img_info(pttrn_rszed_img)
```
