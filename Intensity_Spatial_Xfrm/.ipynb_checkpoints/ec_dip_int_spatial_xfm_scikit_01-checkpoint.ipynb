## Spatial and Intensity Transformation Functions

Image processing tecniques in the spatial domain operate directly on the pixels of an image. In general, we can represent spatial transformations as follows:

<center> $g(x,y) = T[f(x,y)]$ <\center>

where $g(x,y)$ is the result of applying operator $T[\cdot]$ on input image $f(x,y)$.

If we let $(x_0, y_0)$ denote a pixel of interest, spatial transformations operate on a $n \times n$ neighborhood (rectangular region around pixel) of our pixel. When the spatial transformation neighborhood is of size $1 \times 1$ for a grayscale image, our spatial transformation is simply an intensity transformation. Recall that a grayscale image $f(x,y)$ represents intensity values at each pixel $(x,y)$. Thus, an operator that modifies each individual's pixel intensity is simply an intensity operator/transformation.

### Logarithmic and Contrast-Stretching Transformations

The intensity transformations treated in this section are tools for ***dynamic range*** manipulation. With natural images, our objective is typically to enhance the visual perception of an image. With medical images, as the mammogram example shown above, the visual perception enhancements are typically linked with being able to extract meaningful information from the image. In the case of the mammogram image, the contrast enhancement operations we performed allowed a clearer visualization of the tissue lesion.

**Logarithm transformations**

<center> $g(x,y) = c \times log(1 + f(x,y))$ <\center>

Where $c$ is a constant and $f(x,y)$ is an image of floating point *dtype*. 
    
The shape of this transformation is shown below. One of the main uses of a logarithmic transformation is in dynamic range compression. For images with large dynamic ranges (e.g., Fourier transforms with dynamic ranges ~$10^6$), using a log transform results in a narrower dynamic range and we avoid drowning out the low intensity pixels. 
    
Typically, once we have applied the log transformation to our image, we rescale the values to the range [0,1] and finally convert to the desired *dtype* for displaying, printing, etc. 

```python
#plotting the shape of the logiartihmic transformation
log_xfm = lambda c, f : c*np.log(1 + f)

inten_in = np.linspace(0,1,endpoint=True) #generate vector of intensity values in range [0,1]
c = 1 #constant in log transformation

fig_log_xfm, ax_log_xfm = plt.subplots(1,1, figsize=(5,5));
ax_log_xfm.plot(inten_in, log_xfm(c, inten_in)); ax_log_xfm.set_title('Logarithmic Transformation');
```

**Contrast-stretching Transformations**

These type of transformations expand a narrow range of intensity values to a wider range. We saw an example of this with the mammogram image above. 

A contrast stretching transformation has the mathematical form:

<center>$\Large s = T(r) = \frac{1}{1+(k/r)^E}$<\center>
    
Where $s$ is the output image's intensity value, $k$ is a constant and $E$ controls the slope of the function as shown below.
    
Let's visualize a couple of contrast-stretching transformations by varying the functional parameters.
   

```python
#Plotting the contrast-stretching transformation
cntrst_strt_xfm = lambda inten, k, slope : 1/(1 + np.power((k/inten),slope))

inten_in = np.linspace(0.01,1,endpoint=True) #generate vector of intensity values in range [0,1]
k = 0.5 #constant in log transformation
E = 20 #slope parameter in contrast stretching transformation

#Let's visualize the effect of the constant k and slope parameters
#k_lst = np.arange(0.1,1,step=0.25)
k_lst = [0.1, 0.35, 0.6, 0.85]
slope_lst = np.arange(10,50,step=10)

fig_cntrst_strt_xfm, ax_cntrst_strt_xfm = plt.subplots(2,2, figsize=(10,10)); ax_cntrst_strt_xfm = ax_cntrst_strt_xfm.ravel()

for idx, (k, slope) in enumerate(zip(k_lst, slope_lst)):
    ax_cntrst_strt_xfm[idx].plot(inten_in, cntrst_strt_xfm(inten_in, k, slope)); 
    ax_cntrst_strt_xfm[idx].set_title('Contrast-stretching Xfm, k=%s, slope=%s' % (str(k),str(slope)));
    ax_cntrst_strt_xfm[idx].set_xlabel('Input Intensity, $r$'); ax_cntrst_strt_xfm[idx].set_ylabel('Output Intensity, $s$');
    ax_cntrst_strt_xfm[idx].axvline(x=k,linestyle='--',c='r'); ax_cntrst_strt_xfm[idx].legend(['Xfm','k'])
```

As seen from the plots above, as our slope parameter grows, its limit approaches a brickwall-like xfm. This limiting function is a ***thresholding function*** (see below).

```python
#plotting the thresholding function
fig_cntrst_strt_xfm2, ax_cntrst_strt_xfm2 = plt.subplots(1,2, figsize=(10,5)); ax_cntrst_strt_xfm2 = ax_cntrst_strt_xfm2.ravel()

k_lst = [0.5, 0.5]
slope_lst = [20,159]
for idx, (k, slope) in enumerate(zip(k_lst, slope_lst)):
    ax_cntrst_strt_xfm2[idx].plot(inten_in, cntrst_strt_xfm(inten_in, k, slope)); 
    ax_cntrst_strt_xfm2[idx].set_xlabel('Input Intensity, $r$'); ax_cntrst_strt_xfm2[idx].set_ylabel('Output Intensity, $s$');
    ax_cntrst_strt_xfm2[idx].axvline(x=k,linestyle='--',c='r'); ax_cntrst_strt_xfm2[idx].legend(['Xfm','k'])
    
ax_cntrst_strt_xfm2[0].set_title('Contrast-stretching Xfm, k=%s, slope=%s' % (str(k_lst[0]),str(slope_lst[0])));
ax_cntrst_strt_xfm2[1].set_title('Thresholding Xfm, in the limit');
```

### Histogram Processing

The histogram of a digital image represents the approximate distribution of intensity values in the image. For a grayscale image with intensity values in the range $[0, G]$, the *unnormalized* histogram is defined as:

<center>$h(r_k) = n_k$</center>

where $r_k$ is the *k*th intensity level in the interval $[0, G]$ and $n_k$ is the number of pixels in the image with intensity $r_k$. This histogram is unnormalized. 

To get a *normalized* histogram, we simply divide by the number of pixels in our image. For an $M \times N$ image, we get:

<center>$p(r_k) = \Large \frac{h(r_k)}{M \times N} = \frac{n_k}{n}$</center>

From a probabilistic point of view, $p(r_k)$ is an estimate of the probability of intensity value $r_k$.

```python
#Let's plot the histogram of an image

#load image from skimage
from skimage import data, color

#load cat image from skimage
img_cat = color.rgb2gray(data.chelsea())

#get image info
ec_img_utils.get_img_info(img_cat); print('\n')

#Let's plot the histogram for our cat image
hist_cat_orig, hist_cat_ctrs = skimage.exposure.histogram(img_cat)

fig_hist, ax_hist = plt.subplots(1,2,figsize=(20,5)); ax_hist = ax_hist.ravel()
ax_hist[0].imshow(img_cat,cmap='gray'); ax_hist[0].set_title('Original Cat Image');
ax_hist[1].plot(hist_cat_ctrs, hist_cat_orig); ax_hist[1].set_title('Cat Image Unnormalized Histogram');
ax_hist[1].set_xlabel('Intensity'); ax_hist[1].set_ylabel('Intensity Count');
```

**Histogram Equalization**

The goal of histogram equalization is typically image enhancement. We think of the intensity values as a random variable and the image's histogram as an estimate of the discrete probability distribution of the intensity values. The *histogram equalization transformation* of an intensity value $r_k$ into a corresponding random intensity value $s_k$ is defined as a scaling constant, $K$, times the *cumulative distribution function* of the RV $r$.

<center>$s_k = K \cdot CDF(r_k)$</center>

We know that the CDF is defined as:

<center>$CDF(r_k) = \sum_{j=0}^{k} p(r_k) $</center>

Thus, our transformation functions is

<center>$s_k = T(r_k) = K \cdot \sum_{j=0}^{k} p(r_k) $</center>

When $r$ is floating point in the range $[0,1]$, we set $K = 1$. For $r$ integer-valued in the range $[0, L-1]$, we set $K = L-1$.

**Example 3.5 from DIPUM - Histogram Equalization**

```python
#load phoenix lander image
img_phnx_lndr = io.imread('./images/phoenix-lander.tif')

#display the phoenix lander image
fig_phnx, ax_phnx = plt.subplots(2,2,figsize=(10,10)); ax_phnx = ax_phnx.ravel()
ax_phnx[0].imshow(img_phnx_lndr,cmap='gray'); ax_phnx[0].set_title('Phoenix Lander Original');

#perform histogram equalization
img_phnx_eq = skimage.img_as_ubyte( skimage.exposure.equalize_hist(img_phnx_lndr, nbins=256))


#display hist equalized image
ax_phnx[1].imshow(img_phnx_eq,cmap='gray'); ax_phnx[1].set_title('Phoenix Lander Hist Equalized');

#display histograms
hist_phx_orig, hist_phx_ctrs = skimage.exposure.histogram(img_phnx_lndr)
hist_phx_eq, hist_phx_eq_ctrs = skimage.exposure.histogram(img_phnx_eq)

ax_phnx[2].plot(hist_phx_ctrs,hist_phx_orig); ax_phnx[2].set_title('Phoenix Lander Original Hist');
ax_phnx[3].plot(hist_phx_eq_ctrs,hist_phx_eq); ax_phnx[3].set_title('Phoenix Lander Equalized Hist');
```

Note how histogram equalization brings out details that were not visible in the original image. We can verify that the intensity distribution is more evenly distributed in the [0,256] range from the histogram plots shown above.

Let's plot the transformation function corresponding to this histogram equalization transformation.

```python
#Compute normalized histogram 
hist_phx_norm = hist_phx_orig / (img_phnx_lndr.shape[0]*img_phnx_lndr.shape[1])
#print(hist_phx_norm[0:10])

#compute CDF as the cumulative sum
phx_cdf = np.cumsum(hist_phx_norm)

#plot the CDF
fig_cdf, ax_cdf = plt.subplots(1,1)
ax_cdf.plot(np.linspace(0,1,num=256,endpoint=True), phx_cdf)
ax_cdf.set_title('Histogram Equalization Transformation Function'); ax_cdf.set_xlabel('Input intensity values');
ax_cdf.set_ylabel('Output intensity values');
```

**Histogram Matching**

Histogram matching is useful when we need to 'calibrate' an image's histogram against another's. For example, in medical imaging, we may have a 'golden' image that is used to transform subsequent images. 

**Example 3.6 from DIPUM - Histogram Matching**

```python
#load dark X-ray image
img_xray_dark = io.imread('./images/chestXray-dark.tif')

#compute histogram
hist_xray_dark, hist_xray_dark_ctrs = skimage.exposure.histogram(img_xray_dark,normalize=True) 

#load golden X-ray image
img_xray_golden = io.imread('./images/chestXray2-cropped.tif')

#display golden image
fig_golden, ax_golden = plt.subplots(1,2,figsize=(10,5)); ax_golden = ax_golden.ravel()
ax_golden[0].imshow(img_xray_golden,cmap='gray'); ax_golden[0].set_title('Golden X-ray Image')

#compute golden image's histogram and plot
hist_gold, hist_gold_ctr = skimage.exposure.histogram(img_xray_golden,normalize=True)
ax_golden[1].plot(hist_gold_ctr,hist_gold); ax_golden[1].set_title('Golden Histogram')

#compute histogram for golden image
hist_xray_golden, hist_xray_golden_ctrs = skimage.exposure.histogram(img_xray_golden,normalize=True) 
fig_golden.subplots_adjust(wspace = 0.4)
fig_golden.subplots_adjust(hspace = 0.4)

#apply histogram matching xfm
img_xray_hist_match = skimage.exposure.match_histograms(img_xray_dark, reference = img_xray_golden)

#compute histogram of processed image
hist_xray_mtchd, hist_xray_mtchd_ctrs = skimage.exposure.histogram(img_xray_hist_match)

#display both images and their histograms
fig_hist_match, ax_hist_match = plt.subplots(2,2,figsize=(10,10)); ax_hist_match = ax_hist_match.ravel()

ax_hist_match[0].imshow(img_xray_dark,cmap='gray'); ax_hist_match[0].set_title('Dark X-ray image');
ax_hist_match[1].plot(hist_xray_dark_ctrs, hist_xray_dark); ax_hist_match[1].set_xlabel('Intensity'); #ax_hist_match[1].set_ylabel('Intensity Count');
ax_hist_match[1].set_title('Original Histogram');


ax_hist_match[2].imshow(img_xray_hist_match,cmap='gray'); ax_hist_match[2].set_title('Image processed to match golden hist');
ax_hist_match[3].plot(hist_xray_mtchd_ctrs, hist_xray_mtchd); ax_hist_match[3].set_xlabel('Intensity'); #ax_hist_match[3].set_ylabel('Intensity Count');
ax_hist_match[3].set_title('Matched Histogram');

fig_hist_match.subplots_adjust(wspace = 0.4)
fig_hist_match.subplots_adjust(hspace = 0.4)
```

Note from the matched histogram plot shown above that our processed image's intensity distribution has shifted according to our golden histogram's skewness, if you will. 

**Adaptive Histogram Equalization**

Unlike the histogram equalization techniques presented above we transform the intensity values according to the intensity CDF, in adaptive histogram equalization, the image is processed in small regions or tiles. Histogram equalization is then performed on each of the tiles and combined using interpolation to diminish tile boundary artifacts. Tiles with relatively uniform intensity are not equalized as much, and thus, this technique reduces potential noise amplification.

Scikit-image implements Contrast Limited Adaptive Histogram Equalization (CLAHE) where we specify a clip limit in the range [0,1]. This corresponds to histogram matching with a uniform distribution. In Matlab's Image Processing Toolbox, it is possible to specify different histogram distributions (e.g., uniform, rayleigh or exponential) for histogram matching on the tiles.  

**Eample 3.8 from DIPUM - Adaptive histogram equalization**

```python
#Load hidden symbols image
img_hidden_sym = io.imread('./images/hidden-symbols.tif')

#print image information
ec_img_utils.get_img_info(img_hidden_sym); print('\n')

#display original image
fig_adpt, ax_adpt = plt.subplots(1,3,figsize=(15,5))
ax_adpt = ax_adpt.ravel()

ax_adpt[0].imshow(img_hidden_sym,cmap='gray'); ax_adpt[0].set_title('Original Image');

#perform histogram equalization first
img_hidden_hist_eq = skimage.exposure.equalize_hist(img_hidden_sym)

#display the equalized image
ax_adpt[1].imshow(img_hidden_hist_eq,cmap='gray'); ax_adpt[1].set_title('Result from Histogram Equalization');

#perform adaptive histogram equalization
img_hidden_adpt_eq = skimage.exposure.equalize_adapthist(img_hidden_sym, kernel_size=12,clip_limit=0.4)

#display the CLAHE resulting image
ax_adpt[2].imshow(img_hidden_adpt_eq,cmap='gray'); ax_adpt[2].set_title('Result from Adaptive Histogram Equalization');
```

Note how our CDF-based histogram equalization technique was not able to bring out the hidden details in our image. This is due to the algoritum operating on the entire image or on a global scale. On the other hand, when we use adaptive histogram equalization as a local transformation over image tiles, we can bring out the hidden details.

### Linear Spatial Filtering

In this section, we look at spatial filtering where our image transformations consist of multiplying each pixel in a specified neighborhood by a kernel function and summing up the response. That is, for each pixel location $(x,y)$ in our image, our transformation kernel is convolved with a specified neighborhood to transform our intensity values. This type of filtering is different from frequency domain filtering, which we will cover in a separate chapter. 

Some terminology...

***Correlation:*** As in the probabilistic sense, correlation entails sliding or kernel $w$ over an image array so that each element of the kenerl overlays each pixel in the image once. 

***Convolution:*** the convolution process is exactly the same as the correlation process described above, except, our kernel is rotated by $180\deg$ prior to sliding it over each pixel in our image.

Similar to the 1D convolution operation, we can define our two-dimensional convolution operation as follows:

\begin{equation} (w*f)(x,y) = \sum_{s=-a}^{a} \sum_{t=-b}^{b} w(s,t)f(x - s, y - t) \end{equation}

Where $w$ and $f$ are our kernel and image respectively. Constants $a$ and $b$ are given by $a = (m -1)/2$ and $b = (n - 1)/2$ for an $m \times n$ kernel size. 

Convolution is commutative whereas correlation is not. This result is useful and important becase applying multiple linear spatial filters to our image can be thought of as a sequence of convolutions. 

See DIP and DIPUM for diagrams and more details.

**Example 3.9 from DIPUM - averaging kernel**

In scikit-image, the function *skimage.filters.correlate_sparse* performs cross-correlation between an input image and kernel. The different types of modes used to accomodate image padding are outlined in: https://docs.scipy.org/doc/scipy/reference/generated/scipy.ndimage.correlate.html#scipy.ndimage.correlate

In this example, we will be filtering our image with a 31x31 kernel of all 1s. This operation is equivalent to adding up all the pixels in the 31x31 neighborhood.

```python
from skimage import filters

#scikit-image is acting funny when trying to import the correlate_sparse filter. Let's import the correlate function from scipy.ndimage
from scipy.ndimage import correlate
from scipy.ndimage import convolve

#load four-squares image
img_four_sq = io.imread('./images/four-squares.tif')

#display original image
fig_sq, ax_sq = plt.subplots(2,3,figsize=(15,10))
ax_sq = ax_sq.ravel()

ax_sq[0].imshow(img_four_sq,cmap='gray'); ax_sq[0].set_title('Original Image');

#create kernel
w_sq = np.ones((31,31))/(31*31)

#filter image with average kernel and zero-padding
img_sq_avg1 = correlate(img_four_sq,weights=w_sq,mode='constant',cval=0)
#img_sq_avg1 = convolve(img_four_sq,weights=w_sq,mode='constant',cval=0)

#display the filtered image
ax_sq[1].imshow(img_sq_avg1,cmap='gray'); ax_sq[1].set_title('Zero-padding');

#Using the nearest mode which pads the image by extending the nearest pixel - we should not see black borders on the outer perimeter
img_sq_nearest = correlate(img_four_sq,weights=w_sq,mode='nearest')

#display the filtered image
ax_sq[2].imshow(img_sq_nearest,cmap='gray'); ax_sq[2].set_title('Nearest padding');

#Using the reflect mode which pads the image by reflecting about the edfe of the last pixel - we should not see black borders on the outer perimeter
img_sq_reflect = correlate(img_four_sq,weights=w_sq,mode='reflect')

#display the filtered image
ax_sq[3].imshow(img_sq_reflect,cmap='gray'); ax_sq[3].set_title('Reflect padding');

#Using the wrap mode which pads the image by wrapping around the opposite edge - we should see black borders on the outer perimeter
img_sq_wrap = correlate(img_four_sq,weights=w_sq,mode='wrap')

#display the filtered image
ax_sq[4].imshow(img_sq_wrap,cmap='gray'); ax_sq[4].set_title('Wrap padding');


#What happens when we overflow our uint8 image?
img_sq_ovrflw = correlate(img_four_sq,weights=np.ones((31,31)),mode='nearest')

#display the filtered image
ax_sq[5].imshow(skimage.img_as_ubyte(img_sq_ovrflw/np.max(img_sq_ovrflw)),cmap='gray'); ax_sq[5].set_title('Overflowed uint8 Image');
```

**Separable Kernels**

A two-dimensional function $G(x,y)$ is said to be separable if it can be expressed as: $G(x,y) = G_1(x,y)G_2(x,y)$. 

A separable kernel is a matrix that can be expressed as the outer product of of two vectors. For example, 

\begin{equation} w = \begin{bmatrix}
1 & 1 & 1\\
1 & 1 & 1
\end{bmatrix} \end{equation}

is separable because it can be expressed as the outer product: 

\begin{equation} \textbf{c}\textbf{r}^T = \begin{bmatrix} 1 \\ 1 \end{bmatrix} \begin{bmatrix} 1 & 1 & 1 \end{bmatrix} \end{equation}

The outer product of a column vector and a row vector is equivalent to the 2D convolution [DIP], and thus, we are interested in kernels that can be expressed as $w = w_1*w_2$. In the studies that follow, only Gaussian kernels are actually separable. The advantage of separable kernels stem from the fact that convolving an image with separable kernels of size $n \times 1$ as opposed to $n \times n$ or $m \times n$ in the general case is much more computationally efficient. 

If we want to determine whether a kernel or matrix is separable, we first find its rank. If the rank equals 1, then our kernel is separable. This is due to the fact that the outer product of a column and row vector result in a matrix with columns that are linear combinations of one another. 

**Some special kernels and their use/properties**

- *Averaging kernel or box kernel*
- *Disk kernel:* another type of averaging kernel with a circular shape
- *Gaussian lowpass filter kernel*
- *Laplacian kernel:* useful for image sharpening and driving constant areas in an image to zero
- *Log of Laplacian (LoG)* 
- *Gradient kernel:* points in the direction of the greatest rate of change on an image at each pixel location
- *Prewitt and Sobel kernels:* these are two types of gradient kernel implementations

**Example 3.10 from DIPUM - Image smoothing**

```python
#load test pattern image
img_tst_pttrn = io.imread('./images/testpattern1024.tif')

#display original image
fig_pttrn, ax_pttrn = plt.subplots(1,4,figsize=(20,5))
ax_pttrn = ax_pttrn.ravel()

ax_pttrn[0].imshow(img_tst_pttrn, cmap = 'gray'); ax_pttrn[0].set_title('Original Image');

#apply smoothing filter -- Gaussian kernel
#We will make our Gaussian kernel's std dev a percentage of the image's dimensions

#Gaussian lowpass filter
img_pttrn_shape = np.asarray(img_tst_pttrn.shape)

#We will do mild (std dev = 0.25% of image's dim), medium (1%), and extreme (5%) blurring
sig_vect = [0.0025*img_pttrn_shape, 0.01*img_pttrn_shape, 0.05*img_pttrn_shape]

for i in range(len(sig_vect)):
    #filter image and truncate Gaussian tails 6 std dev out
    img_pttrn_gauss1 = filters.gaussian(img_tst_pttrn, sigma = sig_vect[i], mode = 'nearest', truncate = 6.0)

    #display the filtered image
    ax_pttrn[i+1].imshow(img_pttrn_gauss1, cmap = 'gray'); ax_pttrn[i+1].set_title('Gaussian(0, %.2f) LPF'%sig_vect[i][0]);
```

Note how successive increases in the Gaussian kernel's standard deviation increased the smoothing/blurring of the image.

In this example, we use a Laplacian kernel for image enhancement/sharpening.

```python
#load blurry moon picture
img_moon = io.imread('./images/moon-blurry.tif')

#display original image
fig_moon, ax_moon = plt.subplots(1,3, figsize = (15,5))
ax_moon = ax_moon.ravel()

ax_moon[0].imshow(img_moon, cmap = 'gray'); ax_moon[0].set_title('Original Image');

#apply Laplacian filter - note that the Laplacian filter is also used for edge detection. The sequence of steps we'll do here 
img_moon_lplce = filters.laplace(skimage.img_as_float(img_moon),ksize = 3)

#display filtered image
ax_moon[1].imshow(img_moon_lplce, cmap = 'gray'); ax_moon[1].set_title('Laplacian Filtered Image');

#note that the Laplacian filter operates on a float image - Let's verify the dtype of the output image and its range
ec_img_utils.get_img_info(img_moon_lplce)
print('\n')

#restore the gray tones of our image by ADDING the Laplacian image form the original image - See notes below regarding the sign of the Laplace kernel
img_moon_shrp = skimage.img_as_float(img_moon) + img_moon_lplce

ec_img_utils.get_img_info(img_moon_shrp)

#display sharpened image
ax_moon[2].imshow(img_moon_shrp, cmap = 'gray'); ax_moon[2].set_title('Sharpened Image');
```

**High pass filters from lowpass filters**

We can create a high pass filter from a lowpass filter as follows:

\begin{equation} HP(x,y) = \delta(x,y) - LP(x,y) \end{equation}

Because of the sifting property of the delta function, we can equivalently subtract a lowpass filtered version of the image from the original image.

```python
#load the lowpass kernel
with open('lpkernel1D.txt') as lp1d_file:
    #read fir coeff
    data = lp1d_file.read()
    
    #split string
    data = list(data.split(','))
    
    #convert strings to floats
    wLP1D = [float(coeff) for coeff in data]
    
#convert to numpy array
wLP1D = np.asarray(wLP1D).reshape(-1,1)

#create 2D kernel by computing the outer product - This assumes that our filter kernel is separable, but we know that only Gaussian kernels are separable
wLP2D = wLP1D @ np.transpose(wLP1D)

```

**Unsharp Masking**

Unsharp masking filters work by subtracting a blurred version of the image from the original. This is equivalent to a high pass filter as explained above. The steps involved in unsharp filtering are:

1. Blur a copy of the image
2. Subtract the blurred image from the original --> create mask
3. Add the mask to the original image to restore gray tonality

**Example 3.14 from DIPUM**

```python
#load blurry image
img_girl_blurry = io.imread('./images/girl-blurred.tif')

#display images
fig_girl, ax_girl = plt.subplots(2,3,figsize = (15,10))
ax_girl = ax_girl.ravel()

ax_girl[0].imshow(img_girl_blurry, cmap = 'gray'); ax_girl[0].set_title('Original Image');

#Unsharp mask the image --> Note that the name used in scikit-image is a bit of a misnomer since the resulting image is already HPF'ed... In other
#words, we are not getting the mask back that we later have to add to our original image

unshrp_rad = [1.0, 5.0, 5.0, 5.0]
unshrp_amnt = [1.0, 1.0, 1.5, 2.0]

for i in range(len(unshrp_rad)):
    img_girl_unshrp = filters.unsharp_mask(img_girl_blurry, radius = unshrp_rad[i], amount = unshrp_amnt[i])

    ax_girl[i+1].imshow(img_girl_unshrp, cmap = 'gray'); ax_girl[i+1].set_title('Unsharp mask filter (%.1f,%.1f)'%(unshrp_rad[i],unshrp_amnt[i]));

fig_girl.delaxes(ax_girl[5])
```
